```python
columns=['Day','Date','Comments','Local Lead Observer','OS Remote Early Shift','OS Remote Late Shift','DQS Remote Shift','Calibration','FP Setup','FP On-Call']
columns_time=['','','','','(6:00pm-12:30am MST)','(11:45pm-6:00am MST)','(6:00pm-6:00am MST)','(3:00pm-5:00am MST)','(5:45pm-6:15pm MST)','(all night)']
months=['Jan.','Feb.','Mar.','Apr.','May','Jun.','Jul.','Aug.','Sept.','Oct.','Nov.','Dec.']
nb_days=[31,28,31,30,31,30,31,31,30,31,30,31]
weekdays=['Mon.','Tues.','Wed.','Thur.','Fri.','Sat.','Sun.']
#cycle_weekday=cycle(weekdays)
```

```python
def print_month(month,start_day,nb_months=1,col=columns,time=columns_time,mlist=months,nbdlist=nb_days,weekdays=weekdays):
    """
    This function is designed to help create the wiki 
    entries for the observing schedule.
       month : First month for which you need planning. Use an integral number between 1 and 12. For January, enter 1 etc.
       start_day : First day of the month for which you need planning. Use an integral number between 1 and 7. For Tuesday, enter 2 etc.
       nb_months : Number of months you need planning for.
    """
    # initializing indexes
    k=0
    l=0
    first_line=""
    second_line=""
    total_nb_days=0
    m=0
    n=0
    
    # creating the first line
    while(k<len(col)):
        first_line+="||= **"+col[k]+"** ="
        k+=1
    first_line+="||"
    print(first_line)
    
    # creating the second line
    while(l<len(time)):
        second_line+="||= **"+time[l]+"** ="
        l+=1
    second_line+="||"
    print(second_line)
    
    # creating the first month
    m=0
    d=start_day
    while (m<nb_months):
        # creating the line of a given month
        for i in range(nbdlist[month-1]):
            this_line="||= **"+weekdays[d-1]+"** =||= **"+mlist[month-1]+" "+str(i+1)+"** ="
            for j in range(len(col)-2):
                this_line+="||   "
            this_line+="||"
            print(this_line)
            d = (d + 1) % len(weekdays)
        m+=1
        month+=1
```
